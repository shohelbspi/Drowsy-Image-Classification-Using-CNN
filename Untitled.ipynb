```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data_types = os.listdir('datasets')
```

```python
print(data_types)
```

```python
print('types of data found',len(data_types))
```

```python
datas = []

for item in data_types:
    all_data = os.listdir('datasets' + '/'+ item)
    
    for data in all_data:
        datas.append((item,str('datasets' + '/' + item) + '/' + data))
```

```python
len(datas)
```

```python
data_df = pd.DataFrame(data=datas,columns=['data type','images']) 
```

```python
data_df.head(2)
```

```python
data_df.sample(5)
```

```python
data_count = data_df['data type'].value_counts()
```

```python
data_count
```

```python
import cv2
```

```python
path = 'datasets/'

img_size = 300

images = []
labels = []

for i in data_types:
    data_path = path+str(i)
    filesname = [i for i in os.listdir(data_path)]
    for f in filesname:
        img = cv2.imread(data_path + '/' + f)
        img = cv2.resize(img, (86, 86),cv2.INTER_AREA)
        images.append(img)
        labels.append(i)
```

```python
images
```

```python
labels
```

```python
images = np.array(images)
```

```python
images.shape
```

```python
images = images.astype('float32') /255.0
```

```python
images.shape
```

```python
images
```

```python
 y = data_df['data type'].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
y_label_encoder = LabelEncoder()

y = y_label_encoder.fit_transform(y)
```

```python
len(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(images,y,test_size=0.4,random_state=415)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# model create
model = keras.Sequential(
        [
            keras.Input(shape=(86, 86, 3)),
            layers.Conv2D(75, 3, padding='valid', activation='relu'),
            layers.MaxPooling2D(pool_size=(5, 5)),
            layers.Conv2D(64, 2, padding='valid', activation='relu'),
            layers.MaxPooling2D(pool_size=(2, 2)),
            layers.Conv2D(128, 3, padding='valid', activation='relu'),
            layers.MaxPooling2D(pool_size=(2, 2)),
            layers.Flatten(),
            layers.Dense(64, activation='relu'),
            layers.Dense(1, activation='sigmoid'),
        ]
    )

# check model summarry

print(model.summary())
```

```python
# model compile

model.compile(
        loss='binary_crossentropy',
        optimizer='rmsprop',
        metrics=['accuracy']
    )
```

```python
# model training

model.fit(x_train, y_train, batch_size=32, epochs=30)
```

```python
# model prediction

y_pred = model.predict(x_test)
```
